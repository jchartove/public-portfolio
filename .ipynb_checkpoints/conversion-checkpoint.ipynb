Prompt: You are working for an e-commerce company. The company wants to optimize the conversion rate for its website. Conversion refers to the action the company wants the user to take, which in this case is making a purchase. The conversion rate is simply the number of conversions divided by the total number of visitors the page receives.

We have data about users who hit our site. We know whether they converted or not, as well as some of their basic characteristics such as the country theyâ€™re in, the marketing channel that brought them to the site, their age, whether they are repeat users, and the number of pages visited during that session (as a proxy for site activity/time spent on the site).


```python
#required packages in addition to the insight environment: imbalanced-learn, statsmodels

#get data
import pandas as pd
import io
df = pd.read_csv('conversion_data.csv')
df.head()
import warnings
warnings.filterwarnings('ignore')
```

```python
#set up libraries

from sklearn import preprocessing
import matplotlib.pyplot as plt 
import numpy as np
plt.rc("font", size=14)
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
sns.set(style="white")
sns.set(style="whitegrid", color_codes=True)
```

```python
#age histograms

df_g = df.groupby(by=['converted'])
df_converted = df[df['converted'] == 1]
df_un = df[df['converted']==0]

fig, axs = plt.subplots(nrows=3,sharex=True)
df_un['age'].plot.hist(ax=axs[0],bins=np.arange(17,62,1),title='Unconverted')
df_converted['age'].plot.hist(ax=axs[1],bins=np.arange(17,62,1),title='Converted')
df['age'].plot.hist(ax=axs[2],bins=np.arange(17,62,1),title='All')
axs[2].set_xlabel('Age')
```

```python
#pages visited histograms

fig, axs = plt.subplots(nrows=3,sharex=True)
df_un['total_pages_visited'].plot.hist(ax=axs[0],bins=np.arange(0,30,1),title='Unconverted')
df_converted['total_pages_visited'].plot.hist(ax=axs[1],bins=np.arange(0,30,1),title='Converted')
df['total_pages_visited'].plot.hist(ax=axs[2],bins=np.arange(0,30,1),title='Total')
axs[2].set_xlabel('Total pages visited')
```

```python
#region bar plots (log scale)

df_country = df.groupby('country').count()  #.plot.bar()
df_country_totals = pd.DataFrame(data={'Country': df.country.unique(), 'Totals': df_country['converted']})
df_country_u = df_un.groupby('country').count()
df_country_c = df_converted.groupby('country').count()
df_country_totals['Unconverted'] = df_country_u['converted']
df_country_totals['Converted'] = df_country_c['converted']
df_country_totals[['Converted','Unconverted']].plot.bar(stacked='True',logy=True)
df_country_totals
```

```python
#binomial logistic regression model
#the following modeling technique is based on 
#https://towardsdatascience.com/building-a-logistic-regression-in-python-step-by-step-becd4d56c9c8

#more libraries
import imblearn

#SMOTE (Synthetic Minority Oversampling Technique) is a method to improve model training when one category (ie converted)
#has many more instances than another category (ie unconverted) by generating realistic additional training data
from imblearn.over_sampling import SMOTE
os = SMOTE(random_state=0)

#just in case data cleaning
df = df.dropna()
```

```python
#turn categorical variables into numerical dummy variables for modeling
cat_vars=['country','source']
for var in cat_vars:
    cat_list='var'+'_'+var
    cat_list = pd.get_dummies(df[var], prefix=var)
    df1=df.join(cat_list)
    df=df1
cat_vars=['country','source']
```

```python
df_vars=df.columns.values.tolist()
to_keep=[i for i in df_vars if i not in cat_vars]
df_final=df[to_keep]
```

```python
#train_test_split splits arrays or matrices into random train and test subsets
X = df_final.loc[:, df_final.columns != 'converted']
y = df_final.loc[:, df_final.columns == 'converted']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
#os.fit_sample implements SMOTE and resamples the dataset
columns = X_train.columns
os_data_X,os_data_y=os.fit_sample(X_train, y_train)
os_data_X = pd.DataFrame(data=os_data_X,columns=columns )
os_data_y= pd.DataFrame(data=os_data_y,columns=['converted'])
```

```python
# check out the new oversampled data
print("length of original data is ",len(df))
print("length of oversampled data is ",len(os_data_X))
print("Number of unconverted in original data is ",len(df_un))
print("Number of converted in original data is ",len(df_converted))
print("Number of unconverted in oversampled data",len(os_data_y[os_data_y['converted']==0]))
print("Number of converted in oversampled data",len(os_data_y[os_data_y['converted']==1]))
print("Proportion of unconverted data in original data is ",len(df_un)/len(df))
print("Proportion of converted data in original data is ",len(df_converted)/len(df))
print("Proportion of unconverted data in oversampled data is ",len(os_data_y[os_data_y['converted']==0])/len(os_data_X))
print("Proportion of converted data in oversampled data is ",len(os_data_y[os_data_y['converted']==1])/len(os_data_X))
```

```python
#now it is time to fit the logistic model!

import statsmodels.api as sm
logit_model=sm.Logit(os_data_y,os_data_X)
result=logit_model.fit()
print(result.summary2())
```

```python
# odds ratios and 95% CI
# taken from http://blog.yhat.com/posts/logistic-regression-and-python.html
params = result.params
conf = result.conf_int()
conf['OR'] = params
conf.columns = ['2.5%', '97.5%', 'OR']
print(np.exp(conf))
```


Insights from the logistic coefficients:
* Note the column P>|z|; every column in the data is a significant predictor
* The strongest predictor of non-conversion (based on coefficient) is being from China
* There are less differences between other countries but people from the UK are the most likely to make a purchase, followed by the US, followed by Germany
* Based on odds ratio, the strongest predictor of conversion is total pages visited (the more the better)
* Age is slightly inversely correlated with conversion (odds ratio slightly less than 1) - older people less likely to purchase
* The best source for good conversion raes is SEO, the worst is Direct

```python
#time to find out how good the model is. this code fits a classifier to the data
#the classifier is the same as the above model i think?

from sklearn.linear_model import LogisticRegression
from sklearn import metrics
X_train, X_test, y_train, y_test = train_test_split(os_data_X, os_data_y, test_size=0.3, random_state=0)
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
#i don't actually know what all these settings do, if i had more time i'd look it up

LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True, \
                   intercept_scaling = 1, max_iter = 100, multi_class = 'ovr', n_jobs= 1, \
                   penalty='l2', random_state = None, solver = 'liblinear' , tol = 0.0001, verbose=0, warm_start=False)
```

```python
#the point is that the model performs very well

y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

```python
#format of confusion matrix:
#[true positives    false positives]
#[false negatives   true negatives]

from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

```python
#more metrics: precision, recall, F-measure and support

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
#ROC curve

from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
logit_roc_auc = roc_auc_score(y_test, logreg.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, logreg.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```
